```python
from sklearn import svm , datasets
d=datasets.load_iris()
```

```python
import pandas as pd
df=pd.DataFrame(d.data,columns=d.feature_names)
df['flower']=d.target
df
```

```python
from sklearn.model_selection import train_test_split
j_train,j_test,h_train,h_test= train_test_split(d.data,d.target,test_size=0.2)
```

```python
model=svm.SVC(kernel='rbf',C=30,gamma='auto')
model.fit(j_train,h_train)
model.score(j_test,h_test)
```

### grid searchCV

```python
from sklearn.model_selection import GridSearchCV as gcv

clf=gcv(svm.SVC(gamma='auto'),{
    'C':[1,10,20,50],
    'kernel':['rbf','linear']
},cv=5,return_train_score=False)

clf.fit(d.data,d.target)
clf.cv_results_
```

```python
df=pd.DataFrame(clf.cv_results_)
df
```

```python
df[['param_C','param_kernel','mean_test_score']]
```

```python
from sklearn.model_selection import RandomizedSearchCV as rcv

rs=rcv(svm.SVC(gamma='auto'),{
    'C':[1,10,20,50],
    'kernel':['rbf','linear']
},cv=5,return_train_score=False,n_iter=2)

rs.fit(d.data,d.target)
rs.cv_results_

pd.DataFrame(rs.cv_results_)[['param_C','param_kernel','mean_test_score']]
```

```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

# Define hyperparameters for each model
svc_params = {
    'C': [0.1, 1, 10],  
    'kernel': ['linear', 'rbf'],
    'gamma': ['scale', 'auto'],
    'degree': [3, 4, 5]
}

rf_params = {
    'n_estimators': [100, 300, 500],  
    'max_depth': [None, 10, 20],  
    'min_samples_split': [2, 5, 10],  
    'min_samples_leaf': [1, 2, 4],  
    'bootstrap': [True, False]  
}

logreg_params = {
    'penalty': ['l1', 'l2'],  
    'C': [0.001, 0.01, 0.1, 1, 10, 100],  
    'solver': ['liblinear'],  
    'max_iter': [100, 200, 300]  
}

# Combine into a single dictionary
model_params = {
    'SVC': {'model': SVC(), 'params': svc_params},
    'RandomForestClassifier': {'model': RandomForestClassifier(), 'params': rf_params},
    'LogisticRegression': {'model': LogisticRegression(), 'params': logreg_params}
}

# Print combined dictionary
print(model_params)

```

```python
for mod, mp in model_params.items():
    cl = gcv(mp['model'], mp['params'], cv=5, return_train_score=False)
    cl.fit(d.data, d.target)
    scores.append({
        'model': mod,
        'best_score': cl.best_score_,
        'best_params': cl.best_params_
    })

print(scores)
```

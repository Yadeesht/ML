```python

import sklearn as sk
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import f1_score
from sklearn.preprocessing import LabelEncoder
import math
import joblib as jb
from sklearn.model_selection import train_test_split as tts
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv(r"E:\ml datasets\Melbourne_housing_FULL.csv")
df.head()
```

```python
df.nunique(
)
```

```python
df=df.drop(['Address','Date','Postcode','YearBuilt','Lattitude','Longtitude'],axis='columns')
```

```python
df
```

```python
df.isna().sum()
```

```python
row=['Propertycount','Distance','Bedroom2','Bathroom','Car']
df[row]=df[row].fillna(0)
```

```python
df['Landsize']=df['Landsize'].fillna(df['Landsize'].mean())
df['BuildingArea']=df['Landsize'].fillna(df['BuildingArea'].mean())
```

```python
df.dropna(inplace=True)
```

```python
df=pd.get_dummies(df,drop_first=True)
df=df.astype(int)
```

```python
x_train,x_test,y_train,y_test=tts(df.drop(['Price'],axis='columns'),df.Price,test_size=0.3)
```

```python
mod=LinearRegression().fit(x_train,y_train)
mod.score(x_test,y_test)
```

```python
mod.score(x_train,y_train)
```

```python
from sklearn.linear_model import Lasso
modl=Lasso(alpha=100, max_iter=10,tol=0.1)
```

```python
modl.fit(x_train,y_train)
modl.score(x_test,y_test)
```

```python
from sklearn.linear_model import Ridge
model=Ridge(alpha=50, max_iter=100,tol=0.1)
model.fit(x_train,y_train)
model.score(x_test,y_test)
```
